```python
def get_class_ratio(array):
    zeros = 0
    ones = 0
    for i in array:
        if i==0:
            zeros+=1
        elif i==1:
            ones+=1
        else:
            print(i,' target class value is neither one nor zero')
    return zeros,ones
```

```python
relevant_data = data.iloc[:,[0,2,3,-1]]
relevant_data.head()
```

```python
x = relevant_data.iloc[:,[1,2]]
y = relevant_data.iloc[:,-1]
x.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=69420)
x_train,x_val,y_train,y_val = train_test_split(x_train,y_train,test_size=0.2,random_state=69420)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
y_train = y_train.to_numpy()
x_val_scaled = scaler.fit_transform(x_val)
y_val = y_val.to_numpy()
x_test_scaled = scaler.fit_transform(x_test)
y_test = y_test.to_numpy()
```

```python
print(x_train_scaled.shape,y_train.shape,x_val_scaled.shape,y_val.shape,x_test_scaled.shape,y_test.shape)
```

```python
train_zeros,train_ones = get_class_ratio(y_train)
print('train_zeros =',train_zeros,' train_ones =',train_ones,' train total =',train_zeros+train_ones)

val_zeros,val_ones = get_class_ratio(y_val)
print('val_zeros =',val_zeros,' val_ones =',val_ones,' val total =',val_ones+val_zeros)

test_zeros,test_ones = get_class_ratio(y_test)
print('test_zeros =',test_zeros,' test_ones =',test_ones,' test total =',test_ones+test_zeros)
```

```python
def batch_generator(x,y,batch_size):
    half = batch_size//2
    x_batch = np.empty((batch_size,x.shape[1]),dtype='float32')
    y_batch = np.empty((batch_size),dtype='float32')
    yes_pulsar_indexes = np.where(y[:]==1.)[0]
    no_pulsar_indexes = np.where(y[:]==0.)[0]
    while True:
        np.random.shuffle(yes_pulsar_indexes)
        np.random.shuffle(no_pulsar_indexes)
        x_batch[:half] = x[yes_pulsar_indexes[:half]]
        x_batch[half:] = x[no_pulsar_indexes[half:batch_size]]
        y_batch[:half] = y[yes_pulsar_indexes[:half]]
        y_batch[half:] = y[no_pulsar_indexes[half:batch_size]]
        yield x_batch, y_batch
```

```python
train_batch = batch_generator(x_train_scaled,y_train,batch_size=1000)
val_batch = batch_generator(x_val_scaled,y_val,batch_size=250)
x_train_batch,y_train_batch = next(train_batch)
x_val_batch,y_val_batch = next(val_batch)
print(x_train_batch.shape,y_train_batch.shape,"\n")
print(x_val_batch.shape,y_val_batch.shape)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

svm = SVC(random_state=69420)

kernel = ['linear','rbf','poly','sigmoid']
degree = [i+1 for i in range(6)]
C = [i*0.1 for i in range(1,100)]
gamma = ['scale','auto']

svm_param_grid = [
    {'kernel':kernel,'degree':degree,'C':C,'gamma':gamma}
]

svm_grid_search = GridSearchCV(svm,svm_param_grid,n_jobs=-1,cv=5,
                              scoring='neg_mean_squared_error',
                              return_train_score=True)
svm_grid_search.fit(x_train_batch,y_train_batch)
```

```python
svm_grid_search.best_params_
```

```python
best_svm = svm_grid_search.best_estimator_
best_svm.fit(x_train_batch,y_train_batch)
```

```python
from sklearn.metrics import classification_report

svm_pred = best_svm.predict(x_val_batch)
svm_report = classification_report(y_val_batch,svm_pred)
print(svm_report)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

random_forest = RandomForestClassifier(random_state=69420)

n_estimators = [i for i in range(1,200,10)]
max_features = [i for i in range(2,10,2)]
max_depth = [i for i in range(2,20,2)]
param_grid = [
    {'n_estimators':n_estimators,'max_features':max_features,
    'max_depth':max_depth,'bootstrap':[True],}
]

grid_search = GridSearchCV(random_forest,param_grid,n_jobs=-1,
                          cv=5,scoring='neg_mean_squared_error',
                          return_train_score=True)
grid_search.fit(x_train_batch,y_train_batch)
```

```python
grid_search.best_params_
```

```python
best_rf = grid_search.best_estimator_
best_rf.fit(x_train_batch,y_train_batch)
```

```python
from sklearn.metrics import classification_report

rf_pred = best_rf.predict(x_val_batch)
rf_report = classification_report(y_val_batch,rf_pred)
print(rf_report)
```
